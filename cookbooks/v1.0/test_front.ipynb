## Adding info and plotting pandas

```
from coa.geo import GeoInfo as gi # get info with GeoInfo
g=gi()
g.get_list_field()
```

```
data=g.add_field(input=cf.get(where=['france','germany','spain','it'],what='weekly',option='nonneg',which='deaths',when='01/07/2020',output='pandas'),field='area',geofield='where')  # adding area info for countries
data.tail()
```

```
data['deaths per area']=data['weekly']/data['area']
```

```
cf.plot(input=data,input_field='deaths per area')   # use the coa.display module
```

```
cf.plot(input=data,input_field='deaths per area',title='another title',width_height=[300,300])   # change title and size of plot
```

```
cf.scrollmenu_plot(input=data,input_field='deaths per area')   # version scrollmenu
```

## Maps

```
cf._cocoplot.return_map(data,which_data='deaths per area')  # specific field map
```
